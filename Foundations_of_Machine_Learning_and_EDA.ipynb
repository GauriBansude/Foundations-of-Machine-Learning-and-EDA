2.  Explain overfitting and underfitting in ML. How can you detect and prevent them?
- Overfitting: Overfitting occurs when a model learns the training data too well, including noise, and performs poorly on new data.

- Example:

  Training accuracy = 98%

  Testing accuracy = 60%

- Causes

  Too complex model

  Too many features

  Small dataset

- Prevention

  - Cross Validation

  - Regularization (L1/L2)

  - Dropout (DL models)

  - Reduce model complexity

  - More training data

- Underfitting: Underfitting occurs when the model is too simple to capture patterns in data.

- Example:

  Training accuracy = 60%

  Testing accuracy = 58%

- Causes

  Very simple model

  Too few features

  Insufficient training

- Prevention

  - Use a more complex model

  - Increase features

  - Train longer

- Bias-Variance Tradeoff

| Concept       | Meaning      |
| ------------- | ------------ |
| High Bias     | Underfitting |
| High Variance | Overfitting  |

4. What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).
- An imbalanced dataset occurs when one class has significantly more samples than another.
- Random Oversampling: Increase minority samples.
  - Advantage: Balanced dataset

  eg. from imblearn.over_sampling import RandomOverSampler
      ros = RandomOverSampler()
      X_res, y_res = ros.fit_resample(X, y)

  - Disadvantage: Overfitting risk
- SMOTE: Synthetic Minority Over-sampling Technique creates new synthetic samples.
  - Advantages: Better generalization

  eg. from imblearn.over_sampling import SMOTE
      sm = SMOTE()
      X_res, y_res = sm.fit_resample(X, y)

5. Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.
- Feature scaling standardizes the range of variables.
- Large values dominate algorithms.

  - Important for:

    KNN

    SVM

    Gradient Descent
- Min-Max Scaling: Transforms data between 0 and 1

- Formula

  x' = (x - xmin)/(xmax - xmin)
- Standardization (Z-Score): Transforms data with mean = 0 and standard deviation = 1

  z = (x - μ) / σ
- Comparison
| Feature               | Min-Max         | Standardization          |
| --------------------- | --------------- | ------------------------ |
| Range                 | 0–1             | Not fixed                |
| Sensitive to outliers | Yes             | Less                     |
| Used in               | Neural Networks | SVM, Logistic Regression |

6. Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
- Label Encoding: Converts categories into numbers.
- eg.
| Color | Encoded |
| ----- | ------- |
| Red   | 0       |
| Blue  | 1       |
| Green | 2       |
- syntax:
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Color'] = le.fit_transform(df['Color'])
- Use when categories have order (ordinal).

- One-Hot Encoding: Creates separate binary columns.
- eg.
| Color | Red | Blue | Green |
| ----- | --- | ---- | ----- |
| Red   | 1   | 0    | 0     |
| Blue  | 0   | 1    | 0     |
- syntax:
pd.get_dummies(df['Color'])
- Use when categories are nominal (no order).


```
#7 Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons? Dataset: https://github.com/MasteriNeuron/datasets.git
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("googleplaystore.csv")

# convert rating column
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# average rating by category
category_rating = df.groupby('Category')['Rating'].mean().sort_values()

print(category_rating)

# visualization
category_rating.plot(kind='bar', figsize=(10,5))
plt.title("Average Ratings by Category")
plt.show()
```

```
#8 Titanic Dataset a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened? b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) andn adults (Age ≥ 18). Did children have a better chance of survival?
import pandas as pd

df = pd.read_csv("titanic.csv")

survival_rate = df.groupby('Pclass')['Survived'].mean()
print(survival_rate)

df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

survival_age = df.groupby('AgeGroup')['Survived'].mean()
print(survival_age)
```

```
#9 Flight Price Prediction Dataset a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window. b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?\
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("flight_price.csv")

plt.scatter(df['days_left'], df['price'])
plt.xlabel("Days Left")
plt.ylabel("Price")
plt.title("Flight Price vs Days Left")
plt.show()

airline_price = df.groupby('airline')['price'].mean().sort_values()
print(airline_price)
```

```
#10 HR Analytics Dataset a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary). b). Are employees with more projects more likely to leave?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("HR_analytics.csv")

corr = df.corr()

sns.heatmap(corr, annot=True)
plt.show()

project_attrition = df.groupby('number_project')['left'].mean()

print(project_attrition)

project_attrition.plot(kind='bar')
plt.title("Attrition vs Number of Projects")
plt.show()
```
